### Create weekday and hour columns from the datetime column

```python
# create new time columns
import pandas as pd

# Replace 'your_dataset.txt' with the path to your actual dataset file
file_path = '/Users/yujinlee/Downloads/combined_training.txt'

# Read the dataset into a DataFrame
df = pd.read_csv(file_path, sep='\t')
```

```python
df.head()
```

```python
df['1'].head()
```

```python
df['datetime'] = pd.to_datetime(df['1'], format='%Y%m%d%H%M%S%f')

# Extract weekday and hour
df['weekday'] = df['datetime'].dt.day_name()  # This will give the name of the day
df['hour'] = df['datetime'].dt.hour  # This will give the hour

# Now df has the original timestamp, the parsed datetime, the weekday, and the hour
df.head()
```

```python
df.to_csv('/Users/yujinlee/downloads/combined_training.txt', sep='\t', index=False)
```
