```python
# Replace 'your_dataset.txt' with the path to your actual dataset file
file_path = '/Users/yujinlee/Downloads/leaderboard.test.data.20130318_20.txt'

# Read the dataset into a DataFrame
df = pd.read_csv(file_path, sep='\t', header=None)

# Group by the 19th column and count the occurrences of each unique value
frequency = df.groupby(18).size().reset_index(name='Frequency').tail()

# Rename the columns for clarity
frequency.columns = ['18th Column Value', 'Frequency']

# Display the group-by attribute and its frequency
print(frequency)
```
